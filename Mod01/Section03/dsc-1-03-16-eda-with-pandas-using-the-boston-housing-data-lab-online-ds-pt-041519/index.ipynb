```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
df = pd.read_csv('train.csv')
df.head()
```

```python
df.describe()
```

```python
import matplotlib.gridspec as gridspec
plt.style.use('ggplot')
new_figure = plt.figure(figsize=(12,12))
gs = gridspec.GridSpec(2,2)

ax1 = new_figure.add_subplot(gs[0,0])
ax2 = new_figure.add_subplot(gs[0,1])
ax3 = new_figure.add_subplot(gs[1,:])

crime_dispersion = df['crim']
ax1.boxplot(crime_dispersion, showfliers=False)
ax2.boxplot(df['ptratio'], showfliers=True)
ax3.boxplot(df['medv'], showfliers=True, vert=False)
ax1.set_title('Crime Statistics')
ax2.set_title('Teacher to Student Ratio')
ax3.set_title('Median Home Values in 10,000s')
plt.show()
```

```python
crime_by_lstat = df.drop(['zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'medv'], axis=1)
crime_by_lstat.plot('lstat', 'crim', kind='scatter')
```

It seems as if there's a correlation between crime rates and the percentage of population with a low social status. 

```python
home_price = int(df['medv'].mean()*10000)
lowest_price = int(df['medv'].min()*10000)
highest_price=int(df['medv'].max()*10000)
print(f'{home_price} is the average price of a home in Boston.')
print(f'{lowest_price} is the lowest priced home and {highest_price} is the highest priced home.')
print('The following four graphs show the price of home compared to neighborhood statistics.')
```

```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,12))
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.set_xlabel('Median Val of Home in $10,000s')
ax1.set_ylabel('Full Val Property Tax Rate per $10,000')
ax1.set_title('Home Value by Property Tax Rate')
ax1.scatter(df['medv'], df['tax'], color='orange')

ax2.set_xlabel('Median Val of Home in $10,000s')
ax2.set_ylabel('Crime Rate Per Capita')
ax2.set_title('Home Value by Crime Rate')
ax2.scatter(df['medv'], df['crim'], color='red')

ax3.set_xlabel('Median Val of Home in $10,000s')
ax3.set_ylabel('Weighted Mean of Distance to Employment Centers')
ax3.set_title('Home Value by Weighted Mean of Distance to Employment Centers')
ax3.scatter(df['medv'], df['dis'], color='green')

ax4.set_xlabel('Median Val of Home in $10,000s')
ax4.set_ylabel('Average Number of Rooms per Home')
ax4.set_title('Home Value by Average Number of Rooms per Home')
ax4.scatter(df['medv'], df['rm'], color='yellow')
```

Figure 1.1 shows most homes, no matter how much they cost pay less than $400 per $10,000 of value.
Figure 1.2 shows the neighborhoods with more expensive homes see a lower crime rate, though there's an uptick around homes around $500,000.
Figure 2.1 shows that most homes have a close proximity to the five main employment centers in Boston, though a general trend towards lower priced homes closest to the employment centers. 
Figure 2.2 shows that the higher the home value, the more rooms, though the average number of rooms is around 6 per home regardless of price. 

# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
