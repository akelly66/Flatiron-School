## Compare subway traffic by day of the week. Display this as a graph.

```python
df.date = pd.to_datetime(df.date)   
df['DOW'] = df.date.dt.day_name()

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
by_dow = df.groupby('DOW').sum()
by_dow.plot(kind="bar")
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
by_dow = by_dow.reset_index()
by_dow.columns
```

```python
by_dow['Weekend'] = by_dow.DOW.map({'Monday': False, 'Tuesday': False, 'Wednesday': False, 'Thursday': False, 'Friday': False, 'Saturday': True, 'Sunday': True})
wknd = by_dow.groupby('Weekend').mean()
wknd.plot(kind='bar')
```
