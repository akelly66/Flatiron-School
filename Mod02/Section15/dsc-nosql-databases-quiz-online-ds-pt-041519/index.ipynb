## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

They serve different types of data. SQL stores predictable data in tables and rows, where NoSQL can hold data with arbitrary lengths and types of data that are related, or massive datasets that can't be stored on a single server, or manage data for intricate relationships.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB is a document store, in that is uses keys for indicators of what information is contained and values for the content. Key names are arbitrary, so you may not find what you're looking for due to mispellings, case-sensitivity, etc. 

Relational databases have predictable information, and though some of the fields may have missing information, each record contains the same data. MongoDB and other document stores can have a variation of data and fields across records that may only have a few things in common. 






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Fault tolerance mitigates the risk of a server containing necesarry information failing and being unable to be recovered. RDDs stores data and backups of data on multiple servers in case one or more servers fail.
